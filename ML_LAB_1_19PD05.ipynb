```python
def SE(x,y,intc,beta):
    return (1./len(x))*(0.5)*sum(y - beta * x - intc)**2

def L1(intc,beta,lam):
    return lam*(np.abs(intc)+np.abs(beta))

def L2(intc,beta,lam):
    return lam*(intc**2 + beta**2)

N = 100
x = np.random.randn(N)
y = 2 * x + np.random.randn(N)

beta_N = 100
beta = np.linspace(-100,100,beta_N)
intc = 0.0

L1_array = np.array([L1(intc,i,lam=30) for i in beta])
L2_array = np.array([L2(intc,i,lam=1) for i in beta])

fig1 = plt.figure()
ax1 = fig1.add_subplot(1,1,1)
ax1.plot(beta,L1_array,label='L1 norm')
ax1.plot(beta,L2_array,label='L2 norm')

plt.legend()
fig1.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

### Since the data consists of 21 features including the target feature ~ price and 2000 rows, we filter the data to the following features and extract 50 random samples from the df.

```python
df_filtered=pd.DataFrame()
```

```python
df_filtered['battery']=df['battery_power']
df_filtered['mem']=df['int_memory']
df_filtered['wgt']=df['mobile_wt']
df_filtered['clock']=df['clock_speed']
df_filtered['speed']=df['four_g']
df_filtered['price']=df['price_range']
```

```python
df_filtered=df_filtered.sample(50)
```

### The following table describes the mean,standard deviation of the data along with the five number summary.

```python
df_filtered.describe()
```

```python
df_filtered.cov()
```

### Covariance indicates the direction of the linear relationship between variables. Correlation on the other hand measures both the strength and direction of the linear relationship between two variables.Covariance can range from negative infinity to positive infinity.

### Here, in our data battery power shows relatively high positive covariance with price and internal memory of the mobile,weight of it are negative covariance.Negative covariance is an indication that the movement in one variable is opposite to the movement of the other variable.

```python
sns.heatmap(df_filtered.corr(), annot = True,cmap="Spectral_r")
```

### Battery power and speed(4-G) are positively correlated with price
### Weight w.r.t speed and clock w.r.t memory(internal) are negatively correlated

```python
df_filtered.kurt()
```

### Kurtosis is a measure of the combined weight of a distribution's tails relative to the center of the distribution
#### Negative values of kurtosis indicate that a distribution is flat and has thin tails and the outlier character of your data is less extreme that expected had the data come from a normal distribution.
